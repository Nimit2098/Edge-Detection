```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Implementing a function for edge detection.
def edge_detection(image_input):
    #sobel filter
    vertical_filter = [[-1,-2,-1],[0,0,0],[1,2,1]]
    horizontal_filter = [[-1,0,1],[-2,0,2],[-1,0,1]]
    
    #grayscaling the input image
    img = image_input.mean(axis=2, keepdims=True)/255.0
    img = np.concatenate([img]*3, axis=2)

    #initializing and extracting the dimentionns of the image
    edges_img = np.zeros_like(img)
    n,m,d = img.shape

    #initializing for loop for every pixel in the image
    for row in range(3,n-2):
        for col in range(3,m-2):
            local_pixels = img[row-1:row+2, col-1:col+2, 0]

            #passing vertical filter through every pixel
            vertical_transformed_pixels = vertical_filter*local_pixels
            vertical_score = vertical_transformed_pixels.sum()/4

            #passing horizontal filter through every pixel
            horizontal_transformed_pixels = horizontal_filter*local_pixels
            horizontal_score = horizontal_transformed_pixels.sum()/4

            edge_score = (vertical_score**2 + horizontal_score**2)**.5
            edges_img[row,col] = [edge_score]*3
    edges_img = edges_img/edges_img.max()
    return edges_img
```

```python
pikachu_image = plt.imread('download.jpg')
plt.imshow(pikachu_image)
```

```python
pikachu_edge = edge_detection(pikachu_image)
plt.imshow(pikachu_edge)
```
